```python
# 32-Add word-length column

train_df["word_length"] = train_df["text"].astype(str).str.split().str.len()
test_df["word_length"] = test_df["text"].astype(str).str.split().str.len()

print("Word-length column added successfully.")

print("\nTrain dataset columns:")
print(train_df.columns.tolist())

print("\nTest dataset columns:")
print(test_df.columns.tolist())

print("\nFirst 5 rows from train dataset with word length:")
display(train_df[["text", "label", "label_name", "char_length", "word_length"]].head())

print("\nFirst 5 rows from test dataset with word length:")
display(test_df[["text", "label", "label_name", "char_length", "word_length"]].head())
```

```python
# 33-Analyse text length summary

combined_df = pd.concat([train_df, test_df], ignore_index=True)

length_summary = pd.DataFrame({
    "split": ["train", "test", "combined"],

    "char_min": [
        train_df["char_length"].min(),
        test_df["char_length"].min(),
        combined_df["char_length"].min()
    ],

    "char_max": [
        train_df["char_length"].max(),
        test_df["char_length"].max(),
        combined_df["char_length"].max()
    ],

    "char_mean": [
        train_df["char_length"].mean(),
        test_df["char_length"].mean(),
        combined_df["char_length"].mean()
    ],

    "char_median": [
        train_df["char_length"].median(),
        test_df["char_length"].median(),
        combined_df["char_length"].median()
    ],

    "word_min": [
        train_df["word_length"].min(),
        test_df["word_length"].min(),
        combined_df["word_length"].min()
    ],

    "word_max": [
        train_df["word_length"].max(),
        test_df["word_length"].max(),
        combined_df["word_length"].max()
    ],

    "word_mean": [
        train_df["word_length"].mean(),
        test_df["word_length"].mean(),
        combined_df["word_length"].mean()
    ],

    "word_median": [
        train_df["word_length"].median(),
        test_df["word_length"].median(),
        combined_df["word_length"].median()
    ]
})

length_summary = length_summary.round(2)

print("Text length summary:")
display(length_summary)
```

```python
# 34-Analyse text length by label

combined_df = pd.concat([train_df, test_df], ignore_index=True)

length_by_label_summary = (
    combined_df
    .groupby("label_name")
    .agg(
        total_examples=("text", "count"),
        char_min=("char_length", "min"),
        char_max=("char_length", "max"),
        char_mean=("char_length", "mean"),
        char_median=("char_length", "median"),
        word_min=("word_length", "min"),
        word_max=("word_length", "max"),
        word_mean=("word_length", "mean"),
        word_median=("word_length", "median")
    )
    .reset_index()
)

length_by_label_summary = length_by_label_summary.round(2)

print("Text length summary by label:")
display(length_by_label_summary)
```

```python
# 35-Analyse text length by split

train_length_by_split = train_df.copy()
train_length_by_split["split"] = "train"

test_length_by_split = test_df.copy()
test_length_by_split["split"] = "test"

split_length_df = pd.concat(
    [train_length_by_split, test_length_by_split],
    ignore_index=True
)

length_by_split_summary = (
    split_length_df
    .groupby("split")
    .agg(
        total_examples=("text", "count"),
        char_min=("char_length", "min"),
        char_max=("char_length", "max"),
        char_mean=("char_length", "mean"),
        char_median=("char_length", "median"),
        word_min=("word_length", "min"),
        word_max=("word_length", "max"),
        word_mean=("word_length", "mean"),
        word_median=("word_length", "median")
    )
    .reset_index()
)

length_by_split_summary = length_by_split_summary.round(2)

print("Text length summary by split:")
display(length_by_split_summary)
```

```python
# 36-Inspect very short prompts

train_short_check = train_df.copy()
train_short_check["split"] = "train"

test_short_check = test_df.copy()
test_short_check["split"] = "test"

short_check_df = pd.concat(
    [train_short_check, test_short_check],
    ignore_index=True
)

very_short_prompts = short_check_df[
    short_check_df["word_length"] <= 3
].sort_values(["word_length", "char_length"])

print("Number of very short prompts with 3 words or fewer:", len(very_short_prompts))

print("\nVery short prompts:")
display(very_short_prompts[["split", "text", "label", "label_name", "char_length", "word_length"]])

print("\nShortest 10 prompts by word length:")
shortest_10_by_word = short_check_df.sort_values(["word_length", "char_length"]).head(10)
display(shortest_10_by_word[["split", "text", "label", "label_name", "char_length", "word_length"]])

print("\nShortest 10 prompts by character length:")
shortest_10_by_char = short_check_df.sort_values(["char_length", "word_length"]).head(10)
display(shortest_10_by_char[["split", "text", "label", "label_name", "char_length", "word_length"]])
```

```python
# 40-Inspect common text patterns

combined_patterns_df = pd.concat([train_df, test_df], ignore_index=True)

# Basic stopwords to remove from common-word analysis
stopwords = {
    "the", "a", "an", "and", "or", "to", "of", "in", "on", "for", "with",
    "is", "are", "was", "were", "be", "been", "being", "i", "you", "your",
    "me", "my", "we", "our", "it", "this", "that", "as", "at", "by", "from",
    "can", "could", "would", "should", "will", "do", "does", "did", "have",
    "has", "had", "not", "but", "if", "then", "than", "so", "about"
}

def get_top_words(dataframe, label_name, top_n=20):
    label_text = " ".join(
        dataframe[dataframe["label_name"] == label_name]["text"].astype(str).str.lower()
    )

    words = re.findall(r"\b[a-zA-Z][a-zA-Z0-9_+-]*\b", label_text)

    filtered_words = [
        word for word in words
        if word not in stopwords and len(word) > 1
    ]

    word_counts = Counter(filtered_words).most_common(top_n)

    return pd.DataFrame(word_counts, columns=["word", "count"])


safe_common_words = get_top_words(combined_patterns_df, "SAFE", top_n=20)
injection_common_words = get_top_words(combined_patterns_df, "INJECTION", top_n=20)

print("Top 20 common words in SAFE prompts:")
display(safe_common_words)

print("\nTop 20 common words in INJECTION prompts:")
display(injection_common_words)


# Inspect selected injection-related text patterns
patterns_to_check = [
    "ignore",
    "forget",
    "previous",
    "instruction",
    "instructions",
    "system",
    "developer",
    "prompt",
    "jailbreak",
    "bypass",
    "sql",
    "c++",
    "translate",
    "act as",
    "output",
    "print"
]

pattern_results = []

for pattern in patterns_to_check:
    safe_count = combined_patterns_df[
        combined_patterns_df["label_name"] == "SAFE"
    ]["text"].astype(str).str.lower().str.contains(pattern, regex=False).sum()

    injection_count = combined_patterns_df[
        combined_patterns_df["label_name"] == "INJECTION"
    ]["text"].astype(str).str.lower().str.contains(pattern, regex=False).sum()

    pattern_results.append({
        "pattern": pattern,
        "safe_count": safe_count,
        "injection_count": injection_count
    })

pattern_counts_summary = pd.DataFrame(pattern_results)

print("\nSelected text pattern counts:")
display(pattern_counts_summary)
```
